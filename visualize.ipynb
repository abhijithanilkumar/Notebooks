```julia
using NetworkLayout
using LightGraphs
using GraphRecipes
using GeometryTypes
using Plots
```

```julia
g = WheelGraph(10)
```

```julia
a = adjacency_matrix(g)
```

```julia
locs = NetworkLayout.SFDP.layout(a, Point2f0,tol=0.1,C=1,K=1,iterations=100)
```

```julia
x = [p[1] for p in locs]
```

```julia
y = [p[2] for p in locs]
```

```julia
graphplot(a, x=x, y=y)
```
